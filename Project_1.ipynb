Filtering out schools with missing values and creating a new and clean dataframe.

```python
missing_values = ["s", "na"]
schools_filtered = pd.read_csv("https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv", na_values = missing_values)
dft = schools_filtered.dropna()
dft
```

Analysis of descriptive statistics for new dataframe "dft".

```python
dft.shape
```

```python
dft.describe()
```

dft.describe() is far more helpful for analysis. dft is sufficiently large enough to work with and has no missing values.

Calculation of mean scores of initial comparison school Icahn Charter School.

```python
icahn = dft[dft['School Name'] == 'Icahn Charter School']
mean_score_icahn = icahn['Mean Score'].mean()
```

Caclulation of mean scores of the whole borough of the Bronx, which includes Icahn Charter School.


```python
bronx_schools = dft[dft['School DBN'].str.contains('X')]
mean_score_bronx = bronx_schools['Mean Score'].mean()
```

Creation of a dataframe including both mean scores.

```python
dfg = pd.DataFrame({'School': ['Icahn Charter School', 'Bronx Borough Average'], 'Mean Score': [mean_score_icahn, mean_score_bronx]})
#dataframe = pd.DataFrame({'xaxis: ['xparameter1', 'xparameter2'], 'yaxis': [yparameter1, yparameter2]})
```

Data visualization of comparison of means scores between Icahn Charter School and borough of the Bronx using matplot library

```python
dfg.plot(x='School', y='Mean Score', kind='bar', legend=False)
plt.ylabel('Mean Score')
plt.xlabel('School')
plt.title('Mean Score Comparison: Icahn Charter School vs. Bronx Borough Average')
plt.show()
```

Data visualization comparison between Mean Scores and School Level using pandas library


```python
mean_scores_school_level = dft.groupby('School Level')['Mean Score'].mean()
mean_scores_school_level.plot(kind='bar')
```

After transforming the non-standard null values into NaN and deleting them from the original "schools" list, our  new list "dft" shrinks from 212331 values to 69643 values. This list is still sufficiently large enough to work with. We can see that the mean score across schools in NYC in the dft dataframe is about 65. The worst performing school averaged a score of 15 while the highest 98.3. This shows a wide disparity in school performance across the city. We can also see that Icahn Charter School performed significantly better than the rest of its borough. We can also see that High Schools have the lowest scores while K-8 and Junior High have the highest scores.
